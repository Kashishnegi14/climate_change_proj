```python
print("\nüîç Missing Values Analysis:")
missing = df.isnull().sum()
missing_pct = (missing / len(df)) * 100
missing_df = pd.DataFrame({
    'Column': missing.index,
    'Missing Count': missing.values,
    'Missing %': missing_pct.values
}).sort_values('Missing Count', ascending=False)

print(missing_df[missing_df['Missing Count'] > 0])

if missing_df['Missing Count'].sum() == 0:
    print("  ‚úì No missing values found!")
else:
    print(f"  ‚ö† Total missing values: {missing_df['Missing Count'].sum():,}")
```

```python
duplicates = df.duplicated().sum()
print(f"\nüîç Duplicate Rows: {duplicates}")
if duplicates > 0:
    print(f"  ‚ö† Removing {duplicates} duplicate rows...")
    df = df.drop_duplicates()
    print(f"  ‚úì Dataset shape after removing duplicates: {df.shape}")
```

```python
print("\nüîç Outlier Detection (IQR Method):")
numeric_cols = df.select_dtypes(include=[np.number]).columns
outlier_summary = []

for col in numeric_cols:
    if col not in ['Year']:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR
        outliers = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
        outlier_summary.append({
            'Column': col,
            'Outliers': outliers,
            'Outlier %': (outliers / len(df)) * 100
        })

outlier_df = pd.DataFrame(outlier_summary).sort_values('Outliers', ascending=False)
print(outlier_df.head(10))
```

```python
print("\n" + "=" * 80)
print("STEP 3: DATA PREPROCESSING")
print("=" * 80)
```

```python
if df.isnull().sum().sum() > 0:
    print("\nüîß Handling missing values...")
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
for col in numeric_cols:
        if df[col].isnull().sum() > 0:
            df[col].fillna(df[col].median(), inplace=True)
```

```python
categorical_cols = df.select_dtypes(include=['object']).columns
for col in categorical_cols:
        if df[col].isnull().sum() > 0:
            df[col].fillna(df[col].mode()[0], inplace=True)
    
print("  ‚úì Missing values handled!")
```

```python
print("\nüîß Handling outliers (capping at 3√óIQR)...")
for col in numeric_cols:
    if col not in ['Year', 'Population']:
        Q1 = df[col].quantile(0.25)
        Q3 = df[col].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - 3 * IQR
        upper_bound = Q3 + 3 * IQR
        df[col] = df[col].clip(lower=lower_bound, upper=upper_bound)

print("  ‚úì Outliers handled!")

print(f"\n‚úì Data preprocessing complete!")
print(f"  - Final dataset shape: {df.shape}")
```

```python
print("\n" + "=" * 80)
print("STEP 6: TEMPORAL ANALYSIS")
print("=" * 80)

if 'Year' in df.columns:
    print("\nüìà Analyzing trends over time...")
```

```python
# Clean column names for easier access
df.columns = df.columns.str.strip().str.replace(' ', '_').str.replace('(', '').str.replace(')', '')

# Now numeric_cols with cleaned names
numeric_cols = ['Avg_Temperature_¬∞C', 'CO2_Emissions_Tons/Capita', 
                'Sea_Level_Rise_mm', 'Rainfall_mm', 'Population', 
                'Renewable_Energy_%', 'Extreme_Weather_Events', 'Forest_Area_%']

# Group by Year and compute mean
yearly_stats = df.groupby('Year').agg({
    col: 'mean' for col in numeric_cols if col != 'Year'
}).reset_index()
```

```python
fig, axes = plt.subplots(2, 2, figsize=(16, 12))
fig.suptitle('Climate Indicators Over Time', fontsize=16, fontweight='bold')
    
metrics = [col for col in numeric_cols if col != 'Year'][:4]
    
for idx, metric in enumerate(metrics):
        ax = axes[idx // 2, idx % 2]
        ax.plot(yearly_stats['Year'], yearly_stats[metric], marker='o', linewidth=2, markersize=6)
        ax.set_xlabel('Year', fontweight='bold')
        ax.set_ylabel(metric, fontweight='bold')
        ax.set_title(f'{metric} Trend', fontweight='bold')
        ax.grid(alpha=0.3)
        
```

```python
z = np.polyfit(yearly_stats['Year'], yearly_stats[metric], 1)
p = np.poly1d(z)
ax.plot(yearly_stats['Year'], p(yearly_stats['Year']), 
                "r--", alpha=0.8, linewidth=2, label=f'Trend (slope={z[0]:.4f})')
ax.legen+d()
    
plt.tight_layout()
plt.savefig('temporal_trends.png', dpi=300, bbox_inches='tight')
print("  ‚úì Saved: temporal_trends.png")
```

```python
print("\nüìä Growth Rates (First vs Last Year):")
first_year = yearly_stats.iloc[0]
last_year = yearly_stats.iloc[-1]
    
for col in metrics:
        change = last_year[col] - first_year[col]
        pct_change = (change / first_year[col]) * 100
        print(f"  {col}: {change:+.2f} ({pct_change:+.1f}%)")
```
